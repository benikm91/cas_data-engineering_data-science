# CAS Data Science - Exemplarische Prüfungsfragen

### Frage 1

Welche der folgenden Algorithmen kann direkt für ein Klassifikationsproblem eingesetzt werden?

**A** : Decision Trees

**B** : Logistic Regression

**C** : PCA

**D** : Lineare Regression

<details>
  <summary>Korrekte Antwort</summary>
  A, B
</details>

### Frage 2

Welche der folgenden Aussagen sind **korrekt**?  

**A** : Die Lineare Regression nimmt das lineare Modell an und kann nur lineare Funktionen vom Input lernen.

**B** : Obwohl die Polynomielle Regression strikt komplexere Funktionen lernen kann als die Lineare Regression, kann die Lineare Regression trotzdem die bessere Modellwahl sein.   

**C** : Hyper-Parameter (z.B. die Regularisierungsstärke) werden durch ein Optimierungsverfahren (z.B. Gradient-Descent) aus  den Trainingsdaten gelernt.

**D** : Nehmen wir unendlich viele Daten an, kann die Lineare Regression mathematisch von den Inputs alle möglichen Funktionen lernen.

<details>
  <summary>Korrekte Antwort</summary>
  A, B
</details>

### Frage 3

Welche der folgenden Aussagen sind **falsch**?

**A** : Formulieren wir ein zu komplexes Modell, so besteht die Gefahr von *Overfitting.*  

**B** : *Overfitting* kommt dadurch zustande, dass das Optimierungsverfahren nicht konvergiert.  

**C** : *Underfitting* ist nicht schlimm.  

**D** : Overfitting wird mittels Berechnung einer Metrik festgestellt, indem wir die Vorhersagen eines trainierten Modells auf ungesehenen Daten mit den tatsächlichen Werten vergleichen.

<details>
  <summary>Korrekte Antwort</summary>
  B, C
</details>
